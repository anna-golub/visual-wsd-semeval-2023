# Preprocess with preprocessing function
--- handles emojis and non-English phrases and transforms them into English text

```python
def preprocess(row):
  tokens = row['tokens']
  if row['non_eng'] == 0:
    return tokens

  res = []
  for token in tokens:
    if re.match(pattern, token) is not None:
      res.append(token)
      continue
    
    if token == 'üúù':
      token_demojized = 'iron_ore'
    elif token == 'ùÑµ':
      token_demojized = 'cut_time'
    elif token == 'üñæ':
      token_demojized = 'frame_with_an_x'
    elif token == 'üñ∏':
      token_demojized = 'optical_disk'
    else:
      token_demojized = emoji.demojize(token, delimiters=("", ""))
    
    if re.match(pattern, token_demojized) is not None:
      res.append(token_demojized)
      continue
    
    token_translated = translator.translate(token).text
    if re.match(pattern, token_translated) is not None:
      res.append(token_translated)
      continue

    print(token)  # neither an emoji nor a non-English phrase 
    res.append(token)

  return res
```

```python
df.insert(3, 'tokens_preprocessed', df.apply(preprocess, axis=1))
```

```python
df[df['non_eng'] == 1]
```

# Extract target and context

Create columns 
* `target_preprocessed` - preprocessed target word / phrase
* `context_preprocessed` - preprocessed context word / phrase

```python
df.insert(4, 'target_preprocessed', df['tokens_preprocessed'].apply(
    lambda x: re.sub('_', ' ', x[0])
))
df.insert(5, 'context_preprocessed', df['tokens_preprocessed'].apply(
    lambda x: re.sub('_', ' ', x[1])
))
df.head()
```

Save dataframe to file

```python
df.drop(['tokens', 'tokens_preprocessed'], axis=1).to_csv('train_df.csv', sep='\t')
```
